### 결합

```python
# 데이터 작업을 할때 종종 DataFrame 간의 또는 Series 간의 결합이 필요하다
# concat(), join(), merge()와 같은 함수들로 결합이 가능하다
```

concat()

```python
# concat() : 가장 간단한 결합 함수로 기준 축을 따라 나란히 결합한다

canadian_youtube = pd.read_csv("./Data/CAvideos.csv")
british_youtube = pd.read_csv("./Data/GBvideos.csv")
```

```python
canadian_youtube.head(3)
```

```python
british_youtube.head(3)
```

```python
# concat을 이용하여 'canadian_youtube' 데이터 프레임과 'british_youtube' 데이터 프레임을 합쳐주었다
# 기준 축은 axis = 0 또는 1로 설정하는데 기본값(default)은 axis=0이다
# axis = 0(행을 따라 = 세로로 연결 = 수직으로 연결) / 1(열을 따라 = 가로로 연결 = 수평으로 연결)

pd.concat([canadian_youtube, british_youtube])
```

```python
# 단 기준 축으로 연결 될때 연결되는 행 또는 열의 수가 같아야 결측값 없이 연결된다

print(canadian_youtube.shape)
print(british_youtube.shape)
# 기본값 axis=0으로 연결될 때는 세로로 연결되므로 열의 수가 같아야 결측값 없이 연결할 수 있다
```

```python
# 이처럼 기준 축을 axis=1로 바꾸면 가로로 연결되므로 행의 개수가 같아야 결측값 없이 연결할 수 있다
# 그러나 위에서 알수 있듯 행의 개수가 canadian_youtube 는 40881개, british_youtube는 38916 이므로 연결은 되나 british_youtube 쪽에서 밑에 모자라는 행의 개수만큰 결측값이 발생한다
pd.concat([canadian_youtube, british_youtube], axis=1)
```

```python
pd.concat([canadian_youtube, british_youtube], axis=1).info()

# 15번째 열부터 null값 발생
```

join()

```python
# join() : 공통된 인덱스를 가진 두개의 DataFrame을 결합할 때 사용하는 함수

# df1.join(df2, how='left', lsuffix='문자열', rsuffix='문자열')
# df1 : join 함수를 적용할 기준 DataFrame
# df2 : join 함수를 적용하여 병할할 다른 DataFrame
# how : 병합 방법
    # left - 왼쪽 데이터프레임 인덱스를 기준으로 병합
    # right - 오른쪽 데이터프레임 인덱스를 기준으로 병합
    # outer - 인덱스 합집합을 기준으로 병합
    # inner - 인덱스 교집합을 기준으로 병합
# suffix : l 또는 r로 왼쪽 또는 오른쪽 데이터 프레임 열의 접미사 >> 동일한 이름의 열이 있는 경우 생길 수 있는 혼란을 방지하고자 사용
```

```python
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
# left 데이터 프레임과 right 데이터 프레임을 오른쪽 데이터 프레임을 기준으로(how=right) 병합함
# 이때 왼쪽 데이터 프레임 열에는 '_CAN' 접미사를 오른쪽 데이터 프레임 열에는 '_UK' 접미사를 붙여서 표기한다
```

merge()

```python
# merge() : join()과 유사한 두개의 데이터 프레임을 병합하는 함수 단, 인덱스가 아닌 특정 열을 기준으로 병합한다

# pd.merge(df1, df2, on='컬럼명', how='inner', suffixes=('문자열','문자열'))
# df1 : join 함수를 적용할 DataFrame 1
# df2 : join 함수를 적용할 DataFrame 2
# how : 병합 방법
    # left - 왼쪽 데이터프레임 열을 기준으로 병합
    # right - 오른쪽 데이터프레임 열을 기준으로 병합
    # outer - 기준 열의 합집합을 기준으로 병합
    # inner - 기준 열의 교집합을 기준으로 병합
# suffix : (df1의 접미사, df2의 접미사) - 데이터 프레임 열의 접미사 >> 동일한 이름의 열이 있는 경우 생길 수 있는 혼란을 방지하고자 사용
# 이외에도 left_on, right_on, left_index, right_index 인수가 존재하나 여기서는 더 자세히 다루지는 않는다
```

```python
# 예제 데이터프레임 생성
df1 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

df2 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K4'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

# 'key' 열을 기준으로 데이터프레임 병합
result = pd.merge(df1, df2, on='key', how='left', suffixes=('_df1', '_df2'))
# suffixes=('_df1', '_df2') 가 적용이 안되는 이유 >> 동일한 이름의 컬럼이 없기 때문이다

result
```

```python
# 예제 데이터프레임 생성
df1 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K3'],
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

df2 = pd.DataFrame({
    'key': ['K0', 'K1', 'K2', 'K4'],
    'A': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

# 'key' 열을 기준으로 데이터프레임 병합
result = pd.merge(df1, df2, on='key', how='left', suffixes=('_df1', '_df2'))
# 동일한 컬럼 A를 만들어 주면 적용되는 것을 알 수 있다

result
```
