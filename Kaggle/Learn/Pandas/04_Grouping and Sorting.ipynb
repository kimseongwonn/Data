### 그룹별 분석

```python
import pandas as pd

reviews = pd.read_csv("./Data/winemag-data-130k-v2.csv", index_col=0)
```

```python
reviews.head()
```

```python
# groupby() : ()안의 기준으로 그룹별로 묶어준다
# groupby를 사용하여 value_counts()와 같은 기능을 구현할 수 있다

# 'points'를 기준으로 그룹화하여(기준의 고유값들이 라벨이 된다) 각 라벨에 해당하는 'points'값들의 수를 count() 함수로 개수를 세준다
reviews.groupby('points').points.count()
```

```python
# 'points'를 기준으로 그룹화하여(기준의 고유값들이 라벨이 된다) 각 라벨에 해당하는 'price'값들의 최소값을 min() 함수로 찾아준다 
reviews.groupby('points').price.min()
```

```python
# 'winery'별로 그룹화하여 (winery 고유값을 라벨로 사용하여) apply()함수로 ()안의 함수를 적용시킨다
reviews.groupby('winery').apply(lambda df: df.title.iloc[0])

# groupby('winery) >> 'winery'의 고유값을 라벨로 사용하여 라벨에 해당하는 데이터를 그룹화한다
# lambda df: df.title.iloc[0] >> 그룹화된 데이터의 'title' 중 0번째 데이터 값을 가져와서 값으로 사용한다
```

```python
# 추가 설명

# groupby('winery)로 그룹화했을때 '1+1=3'을 기준(라벨)으로 그룹화된 데이터 목록은 아래와 같다
reviews[reviews['winery'] == '1+1=3']
# 이 데이터 모음에서 'title'의 0번째 값을 데이터 값으로 가져와서 사용한다
# 따라서 1+1=3 : 1+1=3 NV Rosé Sparkling (Cava) 와 같은 결과가 나타난다
```

```python
# lambda 식에서 iloc[1]로 바꾸면?
# 1번째 'title'값인 '1+1=3 NV Cygnus Brut Nature Reserva Made With'가 데이터 값으로 사용된다
reviews.groupby('winery').apply(lambda df: df.title.iloc[1] if len(df) > 1 else None)

# if len(df) > 1 else None >> 1번째 title 값이 없는 행의 경우(목록의 데이터 개수가 1 이하인 경우) 오류가 발생하는데 이를 방지하기 위해 넣은 코드
```

여러 열을 기준으로 그룹화

```python
# 더욱 정교한 그룹화를 위해 여러 열을 기준으로 그룹화 할 수 있다

reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
# 'country'로 그룹화 된 결과에서 한번더 'province'로 그룹화한다
```

그룹화하여 여러 함수 적용

```python
# 그룹화를 진행한 뒤 agg()를 사용하여 여러 함수를 한번에 적용할 수 있다
# 여러 함수를 한번에 인수로 넣을때는 리스트 형식으로 넣어야한다

reviews.groupby(['country']).price.agg([len, 'min', 'max']).head()

# 'country'를 기준으로 그룹화한 뒤 'price' 값들에 [len, min, max] 함수를 한번에 agg()를 통해 적용한다
```

### 다중 인덱스(Multi-indexes)

```python
# 두가지 열을 기준으로 그룹화하게 되면 앞서 본것과 같이 첫번째 기준열로 그룹화한 결과에 한번더 두번째 기준열로 그룹화하게된다
# 이때 결과적으로 인덱스가 여러개 생기게되는데 이를 다중 인덱스(Multi-indexes)라고 한다

countries_reviewed = reviews.groupby(['country', 'province']).description.agg([len])
countries_reviewed
```

```python
# 앞서 생긴 데이터 프레임의 다중인덱스의 타입 확인
# >> pandas.core.indexes.multi.MultiIndex
mi = countries_reviewed.index
type(mi)
```

```python
# 다중 인덱스의 경우 단일 인덱스에는 없는 계층 구조를 처리하는 여러가지 방법을 가지고 있다
# 또한 인덱스를 통해 값을 접근하려면 두 레이블이 필요하다

# reset_index() : 다중 인덱스를 단일 인덱스로 다시 변환하는 함수

countries_reviewed.reset_index()
# 다시 일반 인덱스(숫자)가 왼쪽에 생기고 다중 인덱스는 열로 변환되어 나타난다
```

### 정렬

```python
# countries_reviewed 데이터 프레임의 경우 값의 순서가 아닌 인덱스 순서로 데이터를 반환한다
# 이때 값을 기준으로 정렬하고자 하면 sort_values() 함수를 사용할 수 있다

countries_reviewed = countries_reviewed.reset_index()
# reset_index() 결과를 원본 데이터에 할당하여 원본데이터 변경

countries_reviewed.sort_values(by='len')
# 'len' 값을 기준으로 정렬하여 결과 반환
```

```python
# sort_values()의 경우 오름차순 정렬이 기본값(default)이며 가장 낮은 값이 먼저 정렬된다
# 내림차순으로 설정하고 싶으면 ascending=False 를 사용하면 된다

countries_reviewed.sort_values(by='len', ascending=False)
```

```python
# 데이터 값으로 정렬된 결과를 보면 인덱스가 뒤죽박죽이 된 것을 알수 있다
# 이때 sort_index()를 사용하면 다시금 인덱스를 기준으로 정렬시킬 수 있다

countries_reviewed.sort_index()
```

여러 기준으로 한번에 정렬시키기

```python
# 정렬도 그룹화와 마찬가지로 여러 기준으로 정렬시킬 수 있다
# 이때 그룹화와 마찬가지로 첫 번째 기준으로 먼저 정렬된 결과값에서 첫 번째 기준이 동일하다면 두 번째 기준을 근거로 정렬시킨다

countries_reviewed.sort_values(by=['country', 'len'])

# 0번째 행과 1번째 행이 첫번째 기준인 'country'가 Argentina로 같아서 두 번째 기준인 'len' 값을 기준으로 오름차순 정렬되었다
```
