# 응급 의료 시설 지역별 수 시각화

```python
import math

num_cols = len(temp_2022_t.columns)
num_rows = math.ceil(num_cols / 3)

fig, axs = plt.subplots(num_rows, 3, figsize=(15, num_rows * 5))

for i, col in enumerate(temp_2022_t.columns):
    row = i // 3
    col_position = i % 3
    ax = axs[row, col_position]
    ax.bar(temp_2022_t.index[5:], temp_2022_t.iloc[5:,i].values)
    ax.set_title(f'{temp_2022_t.iloc[0,i]},{temp_2022_t.iloc[2,i]},{temp_2022_t.iloc[3,i]}')
    ax.tick_params(axis='x', rotation=45)
    
for j in range(i + 1, num_rows * 3):
    fig.delaxes(axs.flatten()[j])
    
plt.tight_layout()
plt.show()
```
