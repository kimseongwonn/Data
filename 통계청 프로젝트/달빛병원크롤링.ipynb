```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import time
import pandas as pd
from urllib.parse import quote
```

```python
# 저장 리스트
name_list = []
adress_list = []

# 웹 드라이버 설정
driver = webdriver.Chrome()
driver.maximize_window()

driver.get(f"https://www.e-gen.or.kr/moonlight/main.do")
time.sleep(5)

for i in range(90):
    name = driver.find_elements(By.CLASS_NAME, 'hos_name')[i].text
    adress = driver.find_elements(By.CLASS_NAME, 'hos_addr')[i].text
    
    name_list.append(name)
    adress_list.append(adress)

print(name_list)
print(adress_list)
```

```python
data = {
    '병원명' : name_list,
    '주소' : adress_list
}
```

```python
달빛병원 = pd.DataFrame(data)
```
