# MCLP vs P-median
- 커버수요지점수, 수요지점과평균거리 를 비교하여 최종 입지 선택
- case1 : MCLP결과가 P-median 보다 커버수요지점수가 많고 수요지점과평균거리가 짧은 경우 MCLP 결과 채택
- case2 : 동일한 결과값은 그대로 사용
- case3 : MCLP결과가 P-median 보다 커버수요지점수가 적거나 수요지점과평균거리가 긴 경우 P-median 결과 채택

```python
condition1 = mclp_res['커버수요지점수'] > pmedian_res['커버수요지점수']
condition2 = mclp_res['수요지점과평균거리'] < pmedian_res['수요지점과평균거리']
condition3 = mclp_res['요양기관명'] == pmedian_res['요양기관명']
```

```python
# mclp 결과 채택
case1 = mclp_res[condition1 & condition2]
```

```python
# 공통 결과 채택
case2 = mclp_res[condition3]
```

```python
# pmedian 결과 채택
case3 = pmedian_res.iloc[[1,3]]
```

```python
fin_res = pd.concat([case1, case2, case3]).sort_index()
fin_res
```

우선 광주 북구를 샘플로 해서 하나만 진행해 보았습니다

- google에서 제공하는 distancematrix API를 사용하여 시도별 최종 선정 병원과 시도별 수요지점까지의 거리와 소요시간을 추출하였습니다.
- 오류 사항
    - mode=driving(자동차)이 오류가나서 transit(대중교통)으로 거리와 시간 산출

mode >> driving 오류남
