```python
import math

num_cols = len(df_2022_t.columns)
num_rows = math.ceil(num_cols / 3)

fig, axs = plt.subplots(num_rows, 3, figsize=(15, num_rows * 5))

for i, col in enumerate(df_2022_t.columns):
    row = i // 3
    col_position = i % 3
    ax = axs[row, col_position]
    ax.bar(df_2022_t.index[4:14], df_2022_t.iloc[4:14,i].values)
    ax.set_title(f'남성 {df_2022_t.iloc[0,i]}_{df_2022_t.iloc[1,i]}')
    ax.tick_params(axis='x', rotation=45)
    
for j in range(i + 1, num_rows * 3):
    fig.delaxes(axs.flatten()[j])
    
plt.tight_layout()
plt.show()
```

```python
import math

num_cols = len(df_2022_t.columns)
num_rows = math.ceil(num_cols / 3)

fig, axs = plt.subplots(num_rows, 3, figsize=(15, num_rows * 5))

for i, col in enumerate(df_2022_t.columns):
    row = i // 3
    col_position = i % 3
    ax = axs[row, col_position]
    ax.bar(df_2022_t.index[15:], df_2022_t.iloc[15:,i].values)
    ax.set_title(f'여성 {df_2022_t.iloc[0,i]}_{df_2022_t.iloc[1,i]}')
    ax.tick_params(axis='x', rotation=45)
    
for j in range(i + 1, num_rows * 3):
    fig.delaxes(axs.flatten()[j])
    
plt.tight_layout()
plt.show()
```
