오류 해결을 위해 시와 구의 정보가 모두 담긴 시군구코드를 생성하여 데이터 병합

```python
df1.columns
```

```python
# df1의 시도코드와 시군구코드를 결합하여 5자리 문자열로 만듦
df1['시군구코드'] = (df1['행정구역시도코드'].astype(str).str.zfill(2) +
                       df1['행정구역시군구코드'].astype(str).str.zfill(3))

# `code` 데이터프레임의 시군구코드를 문자열로 변환
code['시군구코드'] = code['시군구코드'].astype(str)

# `df1`와 `code`를 'combined_code'와 '시군구코드'를 기준으로 병합
merged_df1 = df1.merge(code, left_on='시군구코드', right_on='시군구코드', how='left')

# 필요한 컬럼 선택 및 이름 변경
merged_df1 = merged_df1[['가구일련번호','행정구역시도코드', '행정구역시군구코드', '시군구코드', '성별코드',
                         '만연령','행정구역시도명','시도_시군구']]
merged_df1.rename(columns={'시도_시군구': '행정구역시군구명'}, inplace=True)

# 결과 출력
merged_df1
```
