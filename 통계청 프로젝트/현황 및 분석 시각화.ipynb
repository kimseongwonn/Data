# 소아응급의료시설 부족 시각화

```python
data = {'권역응급의료센터' : 40, '지역응급의료센터' : 126, '지역응급의료기관' : 244, '응급의료시설' : 114, '권역외상센터' : 17, '소아전문응급의료센터' : 8, }

df = pd.DataFrame(data, index=['병원수'])
df = df.T
df
```

```python
df = df.sort_values('병원수', ascending=False).reset_index()
df.columns = ['병원종류', '병원수']
df
```

```python
import plotly.express as px

min_hospital_type = df.loc[df['병원수'].idxmin(), '병원종류']
colors = ['#75aad0' if hospital_type != min_hospital_type else '#f84b1f' for hospital_type in df['병원종류']]

fig = px.bar(df, x='병원종류', y='병원수', title='병원 종류별 병원 수', labels={'x':'병원종류', 'y':'병원수'})

# 소아과 바그래프 색상 차별화
fig.update_traces(marker_color=colors)

# 차트 사이즈
fig.update_layout(width=800, height=600)

fig.show()
```

# 최초 중증도(KTAS) 데이터 불러오기

```python
level = ['레벨1', '레벨2', '레벨3', '레벨4', '레벨5']
age1 = [349, 12079, 39924, 25292, 4223]
age2 = [1563, 10836, 209848, 274290, 31455]
age3 = [898, 6760, 115171, 174500, 25839]
columns = ['1세 미만', '1-9세', '10-19세']
```

```python
df = pd.DataFrame({
    columns[0]: age1,
    columns[1]: age2,
    columns[2]: age3,
}, index=level)
```

```python
df['종합'] = df.sum(axis=1)
df
```

### 1,2,3 레벨 >> 중증 / 4,5 레벨 >> 경증

- 1세 미만 : 중증(63.9%), 경증(36.1%)
- 1-9세 : 중증(42.1%), 경증(57.9%)
- 10-19세 : 중증(38%), 경증(62%)

```python
import plotly.graph_objects as go

# 데이터 준비
labels = df.index
values = df['종합']

# 파이 차트 생성
fig = go.Figure(data=[go.Pie(labels=labels, values=values, sort=False)])

# 차트 레이아웃 설정 (제목 등)
fig.update_layout(title='연령별 KTAS')

# 파이 차트의 부채꼴에 라벨 표시 설정
fig.update_traces(textinfo='label+percent',
                  insidetextorientation='radial',  # 텍스트 방향 설정
                  outsidetextfont=dict(size=12)    # 외부 텍스트 폰트 크기 설정
                 )

# 차트 보이기
fig.show()
```
