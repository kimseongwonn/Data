문자들이 담겨있는 배열 arr가 주어집니다. arr의 원소들을 순서대로 이어 붙인 문자열을 return 하는 solution함수를 작성해 주세요.

제한사항
- 1 ≤ arr의 길이 ≤ 200
- arr의 원소는 전부 알파벳 소문자로 이루어진 길이가 1인 문자열입니다.

입출력 예

입력값

["a","b","c"]

출력값

"abc"

```python
def solution(arr):
    answer = ''.join(arr)
    return answer

solution(["a","b","c"])
```

문자열 my_string과 정수 k가 주어질 때, my_string을 k번 반복한 문자열을 return 하는 solution 함수를 작성해 주세요.

제한사항
- 1 ≤ my_string의 길이 ≤ 100
- my_string은 영소문자로만 이루어져 있습니다.
- 1 ≤ k ≤ 100

입출력 예

입력값

my_string = "string"

k = 3

출력값

"stringstringstring"

```python
def solution(my_string, k):
    answer = my_string * 3
    return answer

solution("string", 3)
```

연산 ⊕는 두 정수에 대한 연산으로 두 정수를 붙여서 쓴 값을 반환합니다. 예를 들면 다음과 같습니다.

12 ⊕ 3 = 123

3 ⊕ 12 = 312

양의 정수 a와 b가 주어졌을 때, a ⊕ b와 b ⊕ a 중 더 큰 값을 return 하는 solution 함수를 완성해 주세요.

단, a ⊕ b와 b ⊕ a가 같다면 a ⊕ b를 return 합니다.

제한사항
- 1 ≤ a, b < 10,000

입출력 예

입력값

a = 9

b = 91

출력값

result = 991

```python
def solution(a, b):
    result = ''

    if int(str(a)+str(b)) >= int(str(b)+str(a)):
        result = int(str(a)+str(b))
    else:
        result = int(str(b)+str(a))

    return (result)

print(solution(9, 91))
print(solution(1,91))
```

다른 방법

```python
# max(), f'' 사용
# max(a, b) >> a와 b 중 더 큰 값을 반환
# 문자열을 비교하는 경우 첫번째 자리부터 비교하여 더 큰 값을 반환 >> max('agd', 'bhy')의 경우 첫번째 자리 a 와 b를 비교하는데 b가 더 크므로(문자의 경우 늦게 나올 수록 크다) 'bhy'가 반환된다.
# 문자열도 숫자형으로 안바꿔도 크기 비교가 가능하다.(***)

def solution(a, b):
    return int(max(f"{a}{b}", f"{b}{a}"))

print(solution(9, 91))
print(solution(1,91))
```
