연산 ⊕는 두 정수에 대한 연산으로 두 정수를 붙여서 쓴 값을 반환합니다. 예를 들면 다음과 같습니다.

12 ⊕ 3 = 123

3 ⊕ 12 = 312

양의 정수 a와 b가 주어졌을 때, a ⊕ b와 2 * a * b 중 더 큰 값을 return하는 solution 함수를 완성해 주세요.

단, a ⊕ b와 2 * a * b가 같으면 a ⊕ b를 return 합니다.

제한사항
- 1 ≤ a, b < 10,000

입출력 예

1번
- a : 2
- b : 91
- result : 364

2번
- a : 91
- b : 2
- result : 912

```python
def solution(a, b):

    answer = max(int(f'{a}' + f'{b}'),(2 * a * b))

    return answer

print(solution(2,91))
print(solution(91,2))
```

if 문 사용하기

```python
def solution(a, b):

    res1 = f'{a}' + f'{b}'
    res2 = 2 * a * b

    if int(res1) >= res2:
        answer = int(res1)
    else:
        answer = res2

    return answer

print(solution(2,91))
print(solution(91,2))
```

정수 num과 n이 매개 변수로 주어질 때, num이 n의 배수이면 1을 return n의 배수가 아니라면 0을 return하도록 solution 함수를 완성해주세요.

제한사항
- 2 ≤ num ≤ 100
- 2 ≤ n ≤ 9

입출력 예

1번
- num : 98
- n : 2
- result : 1

2번
- num : 34
- n : 3
- result : 0


```python
def solution(num, n):
    if num % n == 0:
        answer = 1
    else:
        answer = 0
    return answer

print(solution(98,2))
print(solution(34,3))
```

```python
# bool 자료형 활용...?
def solution(num, n):
    return int(not(num % n))
    # num%n은 int 값이지만, 이를 not() 함수 안에 넣으면 int가 bool로 해석되어서 num%n이 0이면 False로, 0이 아니면 True로 해석

print(solution(98,2))
print(solution(34,3))
```

```python
# bool 자료형 활용 2
def solution(num, n):
    return int(num % n == 0)
    # 위와 마찬가지로 (num % n == 0)이 int 안에 담겨서 (num % n == 0)이 참이면 1 아니면 0으로 반환

print(solution(98,2))
print(solution(34,3))
```

정수 number와 n, m이 주어집니다. number가 n의 배수이면서 m의 배수이면 1을 아니라면 0을 return하도록 solution 함수를 완성해주세요.

제한사항
- 10 ≤ number ≤ 100
- 2 ≤ n, m < 10

입출력 예

1번
- number = 60
- n = 2
- m = 3
- result = 1

2번
- number = 55
- n = 10
- m = 5
- result = 0

```python
def solution(number, n, m):
    return int(number % n == 0 and number % m == 0)

print(solution(60, 2, 3))
print(solution(55, 10, 5))
```

다른 방법

```python
# if 절 한줄로 쓰기
def solution(number, n, m):
    return 1 if number%n==0 and number%m==0 else 0

print(solution(60, 2, 3))
print(solution(55, 10, 5))
```

```python
def solution(number, n, m):

    if number % n == 0 and number % m == 0: # and, & 둘다 가능
        answer = 1
    else:
        answer = 0

    return answer

print(solution(60, 2, 3))
print(solution(55, 10, 5))
```

양의 정수 n이 매개변수로 주어질 때, n이 홀수라면 n 이하의 홀수인 모든 양의 정수의 합을 return 하고 n이 짝수라면 n 이하의 짝수인 모든 양의 정수의 제곱의 합을 return 하는 solution 함수를 작성해 주세요.

제한사항
- 1 ≤ n ≤ 100

입출력 예

1번
- n = 7
- result = 16

2번
- n = 10
- result = 220

```python
def solution(n):

    answer = 0

    if n % 2 == 0:
        for i in range(n+1):
            if i % 2 == 0:
                answer += i**2
            else:
                continue
    else:
        for j in range(n+1):
            if j % 2 != 0:
                answer += j
            else:
                continue

    return answer

print(solution(7))
print(solution(10))
```

또 다른 방법

```python
# range 함수 interval로 홀짝 구분하기

def solution(n):
    answer = 0
    if n%2:
        for i in range(1,n+1,2):
            answer += i
    else:
        for i in range(2,n+1,2):
            answer += i**2
    return answer

# if n % 2는 n을 2로 나눈 나머지가 0이 아니면 참(True), 즉 홀수를 나타내고, 나머지가 0이면 거짓(False), 즉 짝수를 나타냄
# 이것은 파이썬에서 0이 아닌 숫자는 참(True)으로 간주되기 때문에 가능
# range(1, n+1, 2) >> 1부터 n+1까지 2간격으로 진행 [1,3,5,7...n]
# range(2, n+1, 2) >> 2부터 n+1까지 2간격으로 진행 [2,4,6,8...n]

print(solution(7))
print(solution(10))
```

문자열에 따라 다음과 같이 두 수의 크기를 비교하려고 합니다.

- 두 수가 n과 m이라면
    - ">", "=" : n >= m
    - "<", "=" : n <= m
    - ">", "!" : n > m
    - "<", "!" : n < m

두 문자열 ineq와 eq가 주어집니다. ineq는 "<"와 ">"중 하나고, eq는 "="와 "!"중 하나입니다. 그리고 두 정수 n과 m이 주어질 때, n과 m이 ineq와 eq의 조건에 맞으면 1을 아니면 0을 return하도록 solution 함수를 완성해주세요.

제한 사항
- 1 ≤ n, m ≤ 100

입출력 예

1번
- ineq : "<"
- eq : "="
- n : 20
- m : 50
- result : 1

2번
- ineq : ">"
- eq : "!"
- n : 41
- m :  78
- result : 0


```python
def solution(ineq, eq, n, m):

    answer = 0

    if ineq == ">" and eq == "=":
        if n >= m:
            answer = 1
        else:
            answer = 0
    elif ineq == ">" and eq == "!":
        if n > m:
            answer = 1
        else:
            answer = 0
    elif ineq == "<" and eq == "=":
        if n <= m:
            answer = 1
        else:
            answer = 0
    elif ineq == "<" and eq == "!":
        if n < m:
            answer = 1
        else:
            answer = 0

    return answer

print(solution("<","=", 20, 50))
print(solution(">","!", 41, 78))
```

또 다른 방법

```python
def solution(ineq, eq, n, m):

    if ineq == ">":
        if eq == "=":
            answer = int(n >= m)
        elif eq == "!":
            answer = int(n > m)
    elif ineq == "<":
        if eq == "=":
            answer = int(n <= m)
        elif eq == "!":
            answer = int(n < m)

    return answer

print(solution("<","=", 20, 50))
print(solution(">","!", 41, 78))
```

```python
# 딕셔너리 활용
def solution(ineq, eq, n, m):
    answer = 0

    compareChars = {
        ">=": n >= m,
        "<=": n <= m,
        ">!": n > m,
        "<!": n < m
    }
    # key와 value 값으로 호출 >> [ineq + eq]가 ">="이면 compareChars[ineq + eq]가 n >= m로 호출됨

    answer = int(compareChars[ineq + eq]) # int(참 or 거짓) 활용

    return answer

print(solution("<","=", 20, 50))
print(solution(">","!", 41, 78))
```

```python
# eval() 함수 (@@@@@@@)
# eval(문자열) >> 문자열을 실행 시킴

def solution(ineq, eq, n, m):
    return int(eval(str(n)+ineq+eq.replace('!', '')+str(m)))

print(solution("<","=", 20, 50))
print(solution(">","!", 41, 78))

# replace를 사용해서 문자열에 !가 들어갈 경우 ''로 대체시켜 버림
```
