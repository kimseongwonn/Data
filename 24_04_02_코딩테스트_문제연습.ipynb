다음과 같이 출력하도록 코드를 작성해 주세요.

출력값

!@#$%^&*(\\'"<>?:;

포인트 1 문자열을 나타내는 방법은 3가지이다.
- (' ') 작은 따옴표
- (" ") 큰 따옴표
- (''' ''' / """ """) 삼중 따옴표
- 따라서 출력하고자 하는 문자열에 '와 "가 포함되어 있으므로 삼중 따옴표을 사용해야 오류가 발생되지 않는다.

포인트 2 \\의 활용
- 이스케이프 시퀀스(Escape Sequence)의 시작을 나타내는 특수 문자
- 경로 지정 시 사용 (예 : C:\\Users\\Uesr)
- 문자열 리터럴의 일부로 사용 : (') 작은 따옴표나 (") 큰따옴표를 문자열에 포함시키고 싶을 때 \\'로 사용 >> ('이렇게\'포함시키키')
- \\를 문자열에 포함시키기 : \\ 두 번 사용 (자기 자신을 문자열 리터럴의 일부로 사용하기)

```python
print('''!@#$%^&*(\\'"<>?:;''')
```

또 다른 방법

```python
# 역슬래쉬 사용해서 나타내기
print("!@#$%^&*(\\\'\"<>?:;")
```

```python
# r'' : raw string >> 특수 문자가 이스케이프 되지않고 그대로 출력되는 문자열로 취급
print(r'!@#$%^&*(\'"<>?:;')
```

```python
# f'' : f string 은 특수 문자가 이스케이프 된다.
print(f'!@#$%^&*(\'"<>?:;')
```

두 정수 a, b가 주어질 때 다음과 같은 형태의 계산식을 출력하는 코드를 작성해 보세요.

a + b = c

제한사항
- 1 ≤ a, b ≤ 100

입력값

4 5

출력값

4 + 5 = 9

```python
a, b = map(int, input().strip().split(' '))
print(f'{a} + {b} = {a + b}')
```

- f-string
- 문자열 포맷

```python
a, b = map(int, input().strip().split(' '))
print('{} + {} = {}'.format(a,b,a+b))
```

```python
a, b = map(int, input().strip().split(' '))
print('%d + %d = %d'%(a,b,a+b))
```

```python
a, b = map(int, input().strip().split(' '))
print(a, '+', b, '=', a+b)
```

두 개의 문자열 str1, str2가 공백으로 구분되어 입력으로 주어집니다.

입출력 예와 같이 str1과 str2을 이어서 출력하는 코드를 작성해 보세요.

제한사항
- 1 ≤ str1, str2의 길이 ≤ 10

입력값

apple pen

출력값

applepen

```python
str1, str2 = input().strip().split(' ')
print(str1+str2)
```

또 다른 방법

```python
# sep ='' 파라미터 사용하기

str1, str2 = input().strip().split(' ')
print(str1,str2,sep='')
```

```python
# replace() 사용하여 공백 대체하기

print(input().strip().replace(' ', ''))
```

```python
# join() 사용하여 합쳐주기
# ''.join() >> 구분자.join(이어붙일문자열리스트)

print(''.join(input().strip().split(' ')))
```

```python
# join() 2

str1, str2 = input().strip().split(' ')
print("".join([str1, str2]))
```
