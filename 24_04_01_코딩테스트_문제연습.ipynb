영어 알파벳으로 이루어진 문자열 str이 주어집니다.

각 알파벳을 대문자는 소문자로 소문자는 대문자로 변환해서 출력하는 코드를 작성해 보세요.

제한사항
- 1 ≤ str의 길이 ≤ 20
    - str은 알파벳으로 이루어진 문자열입니다.

입력값

aBcDeFg

출력값

AbCdEfG

답1, 답2 가 틀린 이유 >> 인덱스를 기준으로 짝수 인덱스는 대문자, 홀수 인덱스는 소문자로 출력되는 조건은 예시로 주어진 입력값만 해당되는 조건이다.

반례) ABCDE >> abcde가 아닌 AbCdE가 나오므로 틀린 답

```python
# 답1
str = input()
str_list = []

for j in range(len(str)):
    str_list.append(str[j])

print(str)

for i in range(len(str)):
    if i % 2 == 0:
        str_list[i] = str_list[i].upper()
    else:
        str_list[i] = str_list[i].lower()

print(str)

result = ""

for a in range(len(str_list)):
    result += str_list[a]

print(result)
```

```python
# 답2
str = input()
a = ''

for i in range(len(str)):
    if i % 2 == 0:
        result = str[i:i+1].upper()
    elif i % 2 == 1:
        result = str[i:i+1].lower()

    a += result

print(a)
```

답1, 답2가 틀린점을 깨닫고 바로 만든 정답 코드
아쉬운점
- alpabet 리스트에 대문자를 전부 넣을 필요 없었다.
- 문자열을 인덱스를 사용해서 값을 호출해 비교할 필요없이 바로 i in str을 사용해서 비교하는 것이 더 간편하다.
- str_을 사용하여 값을 담고 a에 넣을 필요 없이 a += str[i].upper/lower()을 사용해도 됐다.

```python
str = input()
a = ''
alpabet = "ABCDEFGHIJKLNMOPQRSYUVWXYZ"

for i in range(len(str)):
    if str[i] in alpabet:
        str_ = str[i].lower()
    else:
        str_ = str[i].upper()

    a += str_

print(a)
```

```python
str = input()
a = ''

for i in range(len(str)):
    if str[i] == str[i].upper():
        a += str[i].lower()
    else:
        a += str[i].upper()

print(a)
```

앞서 말한 아쉬운 점을 전부 고친 코드(최종 답안)

```python
str = input()
a = ''

for i in str:
    if i == i.upper():
        a += i.lower()
    else:
        a += i.upper()

print(a)
```
