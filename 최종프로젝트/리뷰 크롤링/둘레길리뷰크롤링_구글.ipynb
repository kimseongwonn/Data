```python

# 결과 저장 리스트
spot_list = []
name_list = []
stars_list = []
review_list = []
        
def review_craw(place_list):
    for place in place_list:
        
        # 웹 드라이버 설정
        driver = webdriver.Chrome()
        driver.maximize_window()

        driver.get(f'https://www.google.com/maps/')
        time.sleep(3)

        # 선택자로 페이지 요소 선택하기(단일)
        # 지도 검색창 요소 설정
        searchbox = driver.find_element(By.XPATH, '//*[@id="searchboxinput"]')
        # 지도 검색창에 검색 장소 하나씩 넣기
        searchbox.send_keys(place)

        # 검색어로 검색한 뒤 엔터 처리
        searchbutton = driver.find_element(By.CLASS_NAME, "mL3xi")
        searchbutton.click()

        time.sleep(3)

        #@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
        # 2가지 케이스가 존재 >> 경우에 따라 버큰 들의 XPATH가 달라짐 따라서 나누어서 진행
        # 1. 검색시 검색 리스트가 나오는 경우
        # 2. 검색시 바로 검색 장소로 연결되는 경우
        
        # 1. 검색시 바로 연결되지 않고 검색 리스트가 나오는 경우 가장 첫번째 장소 클릭하는 코드
        try:
            # 검색 리스트에서 가장 첫번째 결과 버튼
            firstplacebutton = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[1]/div[1]/div[1]/div/a')
            firstplacebutton.click()
            
            time.sleep(1)
            reviewbutton = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[3]/div/div[1]/div/div/div[2]/div[3]/div/div/button[2]')
            reviewbutton.click()
            
            time.sleep(1)
            # 스크롤 되고 나서 기다리는 시간
            SCROLL_PAUSE_TIME = 2

            # 스크롤할 현재 페이지의 전체 높이
            last_height = driver.execute_script("return document.body.scrollHeight")
            # 스크롤할 시도 기록
            number = 0

            while True:
                # 스크롤 시도 +1
                number = number+1

                # 스크롤바 지정
                scroll = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[3]/div/div[1]/div/div/div[3]')
                # 선택된 scroll을 사용하여 5000픽셀만큼 내림
                driver.execute_script('arguments[0].scrollBy(0, 5000);', scroll)

                # 스크롤 내리고 대기
                time.sleep(SCROLL_PAUSE_TIME)

                # 스크롤 전의 페이지 전체 높이 출력 >> 확인용
                print(f'last height: {last_height}')
                # 다시한번 스크롤바 지정
                scroll = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[3]/div/div[1]/div/div/div[3]')
                # 스크롤 후 페이지 전체 높이 업데이트
                new_height = driver.execute_script("return arguments[0].scrollHeight", scroll)
                # 스크롤 후 페이지 전체 높이 확인
                print(f'new height: {new_height}')

                # 최대 스크롤 횟수 지정 >> 정지조건
                if number == 30:
                    break
                
                # 스크롤 후 페이지 전체 높이가 스크롤 전 페이지 전체 높이와 일치 >> 더이상 내릴 곳이 없다 >> 정지 조건
                if new_height == last_height:
                    break
                
                # 스크롤 진핸시 출력 >> 확인용
                print('cont_scroll')
                # 이전 높이 업데이트 코드
                last_height = new_height
                
        except:
            print(F"firstplacebutton을 버튼을 찾지 못했습니다.")
            

        # 2. 검색시 바로 연결되는 장소 리뷰 클릭
        try:
            reviewbutton = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[3]/div/div/button[2]')
            reviewbutton.click()
            
            time.sleep(2)
            # 스크롤 되고 나서 기다리는 시간
            SCROLL_PAUSE_TIME = 2

            # 스크롤할 현재 페이지의 전체 높이
            last_height = driver.execute_script("return document.body.scrollHeight")
            # 스크롤할 시도 기록
            number = 0

            while True:
                # 스크롤 시도 +1
                number = number+1

                # 스크롤바 지정
                scroll = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]')
                # 선택된 scroll을 사용하여 5000픽셀만큼 내림
                driver.execute_script('arguments[0].scrollBy(0, 5000);', scroll)

                # 스크롤 내리고 대기
                time.sleep(SCROLL_PAUSE_TIME)

                # 스크롤 전의 페이지 전체 높이 출력 >> 확인용
                print(f'last height: {last_height}')
                # 다시한번 스크롤바 지정
                scroll = driver.find_element(By.XPATH, '//*[@id="QA0Szd"]/div/div/div[1]/div[2]/div/div[1]/div/div/div[2]')
                # 스크롤 후 페이지 전체 높이 업데이트
                new_height = driver.execute_script("return arguments[0].scrollHeight", scroll)
                # 스크롤 후 페이지 전체 높이 확인
                print(f'new height: {new_height}')

                # 최대 스크롤 횟수 지정 >> 정지조건
                if number == 30:
                    break
                
                # 스크롤 후 페이지 전체 높이가 스크롤 전 페이지 전체 높이와 일치 >> 더이상 내릴 곳이 없다 >> 정지 조건
                if new_height == last_height:
                    break
                
                # 스크롤 진핸시 출력 >> 확인용
                print('continue_scroll')
                # 이전 높이 업데이트 코드
                last_height = new_height
            
        except:
            print(F"리뷰 버튼을 찾지 못했습니다.")
            
        #@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
        # 리뷰 관련 작성자명, 별점 개수, 리뷰 내용은 동일한 CLASS_NAME을 가지므로 함께 진행
        
        # 리뷰 개수 범위 지정을 위한 요소 지정 >> 작성자명의 수
        review_people = driver.find_elements(By.CLASS_NAME, "d4r55 ")
        print('리뷰 개수 :', len(review_people))
        
        # 리뷰 개수만큼 반복문 사용하여 작성자명, 별점 개수, 리뷰 내용 추출
        for i in range(len(review_people)):
            # 작성자명
            name = driver.find_elements(By.CLASS_NAME, "d4r55 ")[i].text
            
            # 별점개수
            elements = driver.find_elements(By.CSS_SELECTOR, ".kvMYJc[role='img']")[i]
            stars = elements.get_attribute('aria-label')
            
            # 리뷰내용 >> 리뷰 없는 경우 존재하므로 있으면 추출하고 없으면 '리뷰 없음'으로 대체
            try:
                review = driver.find_elements(By.CLASS_NAME, "wiI7pd")[i].text
            except:
                review = '리뷰 없음'
            
            # 결과 리스트에 담기
            spot_list.append(place)
            name_list.append(name)
            stars_list.append(stars)
            review_list.append(review)


    driver.quit()
    
    # 결과 리스트로 dict 생성    
    dict_df = {
        'spot' : spot_list,
        'name' : name_list,
        'stars' : stars_list,
        'review' : review_list,
    }

    # dict로 데이터 프레임 생성
    df = pd.DataFrame(dict_df)
    # 데이터 프레임 확인
    print(df)
    # 데이터 프레임 저장
    df.to_csv(f'../data/서울둘레길/구글리뷰/구글지도_리뷰모음_test.csv', index=False)
```
