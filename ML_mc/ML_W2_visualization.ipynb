# **1-1. plot() 기본 예제**

image.png

```python
# 데이터 생성
x = np.linspace(0, 10, 100) # 0~10사이 100등분분
y = 4 + 2 * np.sin(2 * x)

# plot 그리기
fig, ax = plt.subplots()

ax.plot(x, y, linewidth=2.0) # 선두께

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))
# xlim, ylim : x축과 y축의 범위 설정
# xticks, yticks : x축과 y축의 눈금 설정

plt.grid()
# 그래프에 격자선 추가

plt.show()
```

# **1-2. Advertising data에 적용**

```python
# 데이터 로딩

ad = pd.read_csv("./data/Advertising.csv", index_col=0)
```

```python
ad
```

```python
# 데이터 생성
tv = ad["TV"]
radio = ad["Radio"]
news = ad["Newspaper"]
sales = ad["Sales"]

# plot
fig, ax = plt.subplots(figsize=(12,8))

ax.plot(tv, linewidth=1.0)
ax.plot(radio, linewidth=1.0)
ax.plot(news, linewidth=1.0)
ax.plot(sales, linewidth=1.0)

plt.grid()

plt.show()
```

```python
import math 

# plot
num_cols = len(ad.columns)
num_rows = math.ceil(num_cols/3)

fig, axs = plt.subplots(num_rows, 3, figsize=(15,num_rows * 5))

for i, col in enumerate(ad.columns):
    row = i // 3
    col_position = i % 3
    ax = axs[row, col_position]
    ax.plot(ad[col], linewidth = 1.0)
    ax.set_title(col)
    
    plt.grid()

for j in range(i + 1, num_rows * 3):
    fig.delaxes(axs.flatten()[j])
    

plt.tight_layout
plt.show()
```

# **2-1. scatter() 기본 예제**

image.png

```python
# 데이터 생성
np.random.seed(3) # 난수 고정 시켜서 항상 같은 랜덤 분포가 나오게 함
x = 4 + np.random.normal(0, 2, 24) # 정규 분포(가우시안 분포) (평균, 표준편차, 생성할 갯수)
y = 4 + np.random.normal(0, 2, len(x))

# size and color:
sizes = np.random.uniform(15, 80, len(x)) # uniform distribution(균등 분포) 15~80사이 
colors = np.random.uniform(15, 80, len(x))

# plot
fig, ax = plt.subplots()

ax.scatter(x, y, s=sizes, c=colors)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.grid()

plt.show()
```

image.png

```python
# 데이터 생성
np.random.seed(1)
x = 4 + np.random.normal(0, 1.5, 200)

# plot:
fig, ax = plt.subplots()

ax.hist(x, bins=8, linewidth=0.5, edgecolor="white") 
# bins : histogram의 계급 구간 >> 몇단계로 나누어서 히스토그램을 그릴 것인가
# linewidth : 각 bin들의 테두리 선 두께
# edgecolor : 각 bin들의 테두리 선 색상

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 56), yticks=np.linspace(0, 56, 9))

plt.grid()

plt.show()
```

# **2. PairGrid() - Advertising data**

```python
g = sns.PairGrid(ad)  # 여러 개의 변수를 하나로 표현 
g.map_diag(sns.histplot) # 대각선에는 histplot 보여주기 
g.map_offdiag(sns.scatterplot) # 대각선 아닌 것은 scatter plot 보여주기
```

```python
sns.pairplot(ad)
```

### PairGrid와 pairplot의 차이점
- pairplot의 경우 모든 변수 쌍에 대해 기본적인 시각화를 빠르고 간단하게 보고 싶을 때 유용(대각선은 히스토그램, 나머지는 산점도)
- PairGrid의 경우 복잡하고 다양한 시각적 요구 사항이 있는 경우 그리드의 각 셀에 원하는 차트 유형을 정의하고 추가적인 세부 설정 가능

```python
# PairGrid 객체 생성
g = sns.PairGrid(ad)

# 각 셀에 적절한 플롯 추가하기
g.map_diag(sns.kdeplot)  # 대각선 (같은 변수의 히스토그램)
g.map_offdiag(sns.scatterplot, s=0.2, alpha=0.5)  # 비대각선 (다른 변수의 산점도)

# 추가적인 설정 (예: 회귀선 추가)
g.map_offdiag(sns.regplot)

# 서브플롯 사이의 간격 조정
plt.subplots_adjust(top=0.95)
g.fig.suptitle('PairGrid Example')

# 그리기
plt.show()
```
